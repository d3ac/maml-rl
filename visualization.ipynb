```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Train

```python
data = pd.read_excel('train.xlsx')
data = np.array(data)
plt.plot(data)
plt.xlabel('iteration')
plt.ylabel('return')
plt.title('train')
```

```python
data = pd.read_excel('valid.xlsx')
data = np.array(data)
plt.plot(data)
plt.xlabel('iteration')
plt.ylabel('return')
plt.title('valid')
```

```python
data = np.load('output/results.npz')
train = data['train_returns']
test =  data['valid_returns']
print(f'train: {train}, mean:{np.mean(train)}\n')
print(f'test: {test}, mean{np.mean(test)}\n')
```
